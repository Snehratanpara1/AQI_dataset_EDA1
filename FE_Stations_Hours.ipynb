```python
for col in df.select_dtypes(include='number').columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df[col] = df[col].clip(lower, upper)
```

# Droping Columns(Filter Method)

- Dropping O3 because it doesnt affect more to the AQI
- Dropping NOx and PM10 because they are coreelated with NO and PM2.5 respectively

```python
df = df.drop(columns=['O3', 'NOx', 'PM10'])
```

```python
df.sample(5)
```

```python
from sklearn.feature_selection import VarianceThreshold

selector = VarianceThreshold(threshold=0.0)
selector.fit(df.select_dtypes(include='number'))

# See which columns are kept
kept_cols = df.select_dtypes(include='number').columns[selector.get_support()]
print("Kept columns:", kept_cols)

# See which columns are dropped
dropped_cols = df.select_dtypes(include='number').columns[~selector.get_support()]
print("Dropped columns:", dropped_cols)
```

**Emdedding Methods**

```python
from sklearn.ensemble import RandomForestRegressor

X = df.drop(columns=['AQI', 'City', 'State', 'AQI_Bucket', 'Date', 'Time', 'StationName', 'Status','State_encoded','City_encoded'])
y = df['AQI']

rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X, y)

importance = pd.Series(rf.feature_importances_, index=X.columns)
importance.sort_values(ascending=False).plot(kind='bar', figsize=(12,6))
plt.show()
```



# Feature Selection

## Overview
Feature selection was performed using multiple methods to identify 
the most relevant features for AQI prediction.

## Methods Applied

### 1. Filter Method - Correlation Analysis
- Calculated correlation of all pollutants with AQI
- **Dropped O3** → correlation of only -0.14 with AQI (near zero)
- **Dropped NOx** → highly correlated with NO (0.78), redundant
- **Dropped PM10** → highly correlated with PM2.5 (0.90), redundant

### 2. Filter Method - Variance Threshold
- Applied VarianceThreshold with threshold = 0.0
- **No columns dropped** → all features had meaningful variance
- Confirms dataset is rich and varied across cities and time

### 3. Embedded Method - Random Forest Feature Importance
- Trained RandomForestRegressor with 100 estimators
- **Dropped Season columns** → near zero importance (~0.00)
- **Dropped NH3, NO, CO** → near zero importance (~0.00)

## Key Findings
- **PM2.5 is the most important feature** → 92% importance score
- **Location (City, State) strongly influences AQI** → target 
  encoding captured city-wise pollution levels
- **Seasonal features added no value** once pollutants were known
- **Not all pollutants are equal** → only PM2.5, NO2, SO2, 
  Toluene, Benzene showed meaningful importance
