**Filling Missing values according to the city base means Making group of city and filling values with median of there specific city columns**

```python
for col in df.select_dtypes(include='number').columns.drop('AQI'):
    df[col] = df.groupby('City')[col].transform('median')
```

**Finding Outliers**

```python
Q1 = df.select_dtypes(include='number').quantile(0.25)
Q3 = df.select_dtypes(include='number').quantile(0.75)
IQR = Q3 - Q1

outliers = ((df.select_dtypes(include='number') < (Q1 - 1.5 * IQR)) | 
            (df.select_dtypes(include='number') > (Q3 + 1.5 * IQR)))

outliers.sum()
```

```python
(outliers.sum() / len(df) * 100).round(2)
```

```python
len(df)
```

```python
for col in df.select_dtypes(include='number').columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df[col] = df[col].clip(lower, upper)
```

```python
(outliers.sum() / len(df) * 100).round(2)
```
